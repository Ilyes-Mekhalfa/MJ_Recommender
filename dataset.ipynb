```python
import pandas as pd
import random
import json
from datetime import datetime, timedelta

# === ALGERIAN WILAYAS (48 provinces) ===
WILAYAS = [
    "Adrar", "Chlef", "Laghouat", "Oum El Bouaghi", "Batna", "BÃ©jaÃ¯a", "Biskra", "BÃ©char",
    "Blida", "Bouira", "Tamanrasset", "TÃ©bessa", "Tlemcen", "Tiaret", "Tizi Ouzou", "Alger",
    "Djelfa", "Jijel", "SÃ©tif", "SaÃ¯da", "Skikda", "Sidi Bel AbbÃ¨s", "Annaba", "Guelma",
    "Constantine", "MÃ©dÃ©a", "Mostaganem", "M'Sila", "Mascara", "Ouargla", "Oran", "El Bayadh",
    "Illizi", "Bordj Bou Arreridj", "BoumerdÃ¨s", "El Tarf", "Tindouf", "Tissemsilt", "El Oued",
    "Khenchela", "Souk Ahras", "Tipaza", "Mila", "AÃ¯n Defla", "NaÃ¢ma", "AÃ¯n TÃ©mouchent",
    "GhardaÃ¯a", "Relizane"
]

# === INTEREST CATEGORIES ===
INTERESTS = [
    "technology", "AI", "coding", "innovation", "robotics", "cybersecurity",
    "art", "music", "culture", "photography", "painting", "cinema", "theatre",
    "sports", "football", "basketball", "volleyball", "athletics", "swimming",
    "volunteering", "environment", "leadership", "entrepreneurship", "public speaking",
    "literature", "poetry", "writing", "history", "gaming", "esports",
    "dance", "fashion", "design", "cooking", "handicrafts", "calligraphy"
]

# Interest categories for grouping
INTEREST_CATEGORIES = {
    "technology": ["technology", "AI", "coding", "innovation", "robotics", "cybersecurity"],
    "arts": ["art", "music", "culture", "photography", "painting", "cinema", "theatre", "dance", "fashion", "design"],
    "sports": ["sports", "football", "basketball", "volleyball", "athletics", "swimming"],
    "social": ["volunteering", "environment", "leadership", "entrepreneurship", "public speaking"],
    "literature": ["literature", "poetry", "writing", "history"],
    "gaming": ["gaming", "esports"],
    "crafts": ["cooking", "handicrafts", "calligraphy"]
}

# === ALGERIAN NAMES ===
FIRST_NAMES_MALE = [
    "Mohammed", "Ahmed", "Youcef", "Amine", "Karim", "Rami", "Walid", "Bilal", "Hamza", "Sofiane",
    "Mehdi", "Rayan", "Aymen", "Ilyes", "Yacine", "Nassim", "Farid", "Samir", "Djamel", "Nabil",
    "Abdelkader", "Hichem", "Rachid", "Tarek", "Zakaria", "Adel", "Islam", "Mourad", "Hakim", "Nadir"
]

FIRST_NAMES_FEMALE = [
    "Amina", "Sara", "Yasmine", "Nadia", "Lina", "Meriem", "Fatima", "Kenza", "Lamia", "Selma",
    "Asma", "Houria", "Sabrina", "Nesrine", "Ikram", "Sihem", "Imane", "Rania", "Soraya", "Leila",
    "Malika", "Hafida", "Samia", "Zineb", "Wafa", "Rima", "Dounia", "Amel", "Lynda", "Karima"
]

LAST_NAMES = [
    "Bouaziz", "Benamar", "Khaled", "Mansouri", "Saidi", "Hamidi", "Benali", "Meziane", "Lakhal",
    "Cherif", "Amrani", "Boudiaf", "Merad", "Belaidi", "Ziane", "Touati", "Hadji", "Belkacem",
    "Benabdallah", "Bouras", "Khelifa", "Azzouz", "Ould", "Benyoucef", "Ferhat", "Mebarki"
]

# === EVENT TITLES BY CATEGORY ===
EVENT_TEMPLATES = {
    "technology": [
        "Hackathon {theme}", "AI Workshop: {theme}", "Coding Bootcamp {theme}",
        "Tech Talk: {theme}", "Innovation Challenge {theme}", "Robotics Competition {theme}",
        "Cybersecurity Summit {theme}", "Mobile App Dev Workshop", "Web Development Jam"
    ],
    "arts": [
        "Art Exhibition: {theme}", "Music Festival {theme}", "Cultural Week {theme}",
        "Photography Contest {theme}", "Theatre Performance: {theme}", "Cinema Club Screening",
        "Dance Competition {theme}", "Fashion Show {theme}", "Design Workshop {theme}"
    ],
    "sports": [
        "Football Tournament {theme}", "Basketball Championship {theme}", "Volleyball League {theme}",
        "Athletics Meet {theme}", "Swimming Competition {theme}", "Sports Festival {theme}",
        "Marathon for Youth", "3x3 Basketball Challenge", "Futsal Cup"
    ],
    "social": [
        "Leadership Camp {theme}", "Environmental Cleanup {theme}", "Volunteering Day {theme}",
        "Entrepreneurship Workshop {theme}", "Public Speaking Contest {theme}", "Green Challenge {theme}",
        "Community Service Project", "Youth Forum: {theme}", "Social Innovation Lab"
    ],
    "literature": [
        "Poetry Slam {theme}", "Book Club: {theme}", "Writing Workshop {theme}",
        "Literary Debate {theme}", "History Lecture: {theme}", "Author Meetup",
        "Creative Writing Contest", "Storytelling Night", "Reading Marathon"
    ],
    "gaming": [
        "Esports Tournament: {theme}", "Gaming Championship {theme}", "LAN Party {theme}",
        "FIFA Competition", "PUBG Mobile Cup", "Valorant Challenge", "Gaming Jam Session"
    ],
    "crafts": [
        "Cooking Workshop: {theme}", "Handicraft Exhibition {theme}", "Calligraphy Class {theme}",
        "Traditional Crafts Fair", "DIY Maker Space", "Pottery Workshop"
    ]
}

THEMES = [
    "2025", "Youth Edition", "for Beginners", "Advanced", "National", "Regional",
    "Spring Edition", "Summer Special", "Winter Cup", "Annual", "Open", "Elite"
]

# === GENERATE USERS ===
def generate_users(num_users=1000):
    users = []
    for i in range(1, num_users + 1):
        is_male = random.choice([True, False])
        first_name = random.choice(FIRST_NAMES_MALE if is_male else FIRST_NAMES_FEMALE)
        last_name = random.choice(LAST_NAMES)
        wilaya = random.choice(WILAYAS)
        
        # Generate 2-5 interests, with some correlation
        num_interests = random.randint(2, 5)
        category = random.choice(list(INTEREST_CATEGORIES.keys()))
        
        # 70% from same category, 30% from others
        interests = []
        category_interests = INTEREST_CATEGORIES[category]
        interests.extend(random.sample(category_interests, min(num_interests, len(category_interests))))
        
        # Add some random interests
        remaining = num_interests - len(interests)
        if remaining > 0:
            other_interests = [i for i in INTERESTS if i not in interests]
            interests.extend(random.sample(other_interests, min(remaining, len(other_interests))))
        
        users.append({
            "id": i,
            "name": f"{first_name} {last_name}",
            "wilaya": wilaya,
            "interests": interests[:num_interests]
        })
    
    return users

# === GENERATE ANNEXES ===
def generate_annexes():
    annexes = []
    annex_id = 1
    
    for wilaya in WILAYAS:
        # Main annex per wilaya
        annexes.append({
            "annex_id": annex_id,
            "annex_name": f"Maison de Jeunes {wilaya} Centre",
            "wilaya": wilaya,
            "commune": "Centre",
            "activities": random.sample(INTERESTS, random.randint(3, 6))
        })
        annex_id += 1
    
    return annexes

# === GENERATE EVENTS ===
def generate_events(num_events=1000, annexes=[]):
    events = []
    start_date = datetime(2025, 11, 1)
    
    for i in range(1, num_events + 1):
        # Pick a category
        category = random.choice(list(INTEREST_CATEGORIES.keys()))
        available_cats = INTEREST_CATEGORIES[category]
        num_cats = min(random.randint(1, 3), len(available_cats))
        categories = random.sample(available_cats, num_cats)
        
        # Generate title
        template = random.choice(EVENT_TEMPLATES[category])
        theme = random.choice(THEMES)
        title = template.format(theme=theme)
        
        # Random date within next 6 months
        days_ahead = random.randint(1, 180)
        event_date = start_date + timedelta(days=days_ahead)
        
        # Pick random annex
        annex = random.choice(annexes)
        
        events.append({
            "id": i,
            "title": title,
            "category": categories,
            "annex_id": annex["annex_id"],
            "date": event_date.strftime("%Y-%m-%d"),
            "location": annex["wilaya"]
        })
    
    return events

# === GENERATE GROUND TRUTH (REALISTIC INTERACTIONS) ===
def generate_ground_truth(users, events):
    """
    Generate realistic user-event interactions
    Users are more likely to attend events that match their interests and location
    """
    ground_truth = {}
    
    for user in users:
        user_id = user["id"]
        user_interests = set(user["interests"])
        user_wilaya = user["wilaya"]
        
        # Find events that match user interests
        matching_events = []
        for event in events:
            event_categories = set(event["category"])
            
            # Calculate match score
            interest_overlap = len(user_interests & event_categories)
            location_match = 1 if event["location"] == user_wilaya else 0
            
            # Events with good overlap or same location
            if interest_overlap >= 1 or (location_match and random.random() > 0.7):
                score = interest_overlap * 2 + location_match
                matching_events.append((event["id"], score))
        
        # Sort by score and pick top 3-7 events randomly
        matching_events.sort(key=lambda x: x[1], reverse=True)
        num_to_pick = random.randint(3, min(7, len(matching_events)))
        
        if matching_events:
            selected = [event_id for event_id, _ in matching_events[:num_to_pick * 2]]
            ground_truth[user_id] = random.sample(selected, min(num_to_pick, len(selected)))
        else:
            ground_truth[user_id] = []
    
    return ground_truth

# === MAIN GENERATION ===
print("ğŸš€ Generating Large-Scale Algerian Youth Event Dataset...\n")

print("ğŸ“Š Generating users...")
users = generate_users(1000)
print(f"   âœ“ Created {len(users)} users")

print("ğŸ¢ Generating annexes...")
annexes = generate_annexes()
print(f"   âœ“ Created {len(annexes)} annexes (one per wilaya)")

print("ğŸ‰ Generating events...")
events = generate_events(1000, annexes)
print(f"   âœ“ Created {len(events)} events")

print("ğŸ¯ Generating ground truth interactions...")
ground_truth = generate_ground_truth(users, events)
print(f"   âœ“ Created {len(ground_truth)} user interaction records")

# === STATISTICS ===
print("\n" + "="*60)
print("DATASET STATISTICS")
print("="*60)

# User stats
total_interests = sum(len(u["interests"]) for u in users)
print(f"\nğŸ‘¥ Users: {len(users)}")
print(f"   - Average interests per user: {total_interests / len(users):.1f}")
print(f"   - Wilayas represented: {len(set(u['wilaya'] for u in users))}")

# Event stats
total_categories = sum(len(e["category"]) for e in events)
print(f"\nğŸ‰ Events: {len(events)}")
print(f"   - Average categories per event: {total_categories / len(events):.1f}")
print(f"   - Locations: {len(set(e['location'] for e in events))}")

# Ground truth stats
total_interactions = sum(len(gt) for gt in ground_truth.values())
print(f"\nğŸ¯ Ground Truth:")
print(f"   - Total interactions: {total_interactions}")
print(f"   - Average events per user: {total_interactions / len(ground_truth):.1f}")

# Interest distribution
interest_counts = {}
for user in users:
    for interest in user["interests"]:
        interest_counts[interest] = interest_counts.get(interest, 0) + 1

print(f"\nğŸ“Š Top 10 Interests:")
for interest, count in sorted(interest_counts.items(), key=lambda x: x[1], reverse=True)[:10]:
    print(f"   - {interest}: {count} users")

# === SAVE TO FILES ===
print("\nğŸ’¾ Saving datasets...")

# Save as JSON
with open('users_1000.json', 'w', encoding='utf-8') as f:
    json.dump(users, f, ensure_ascii=False, indent=2)

with open('events_1000.json', 'w', encoding='utf-8') as f:
    json.dump(events, f, ensure_ascii=False, indent=2)

with open('annexes_48.json', 'w', encoding='utf-8') as f:
    json.dump(annexes, f, ensure_ascii=False, indent=2)

with open('ground_truth_1000.json', 'w', encoding='utf-8') as f:
    json.dump(ground_truth, f, ensure_ascii=False, indent=2)

# Save as CSV
pd.DataFrame(users).to_csv('users_1000.csv', index=False, encoding='utf-8')
pd.DataFrame(events).to_csv('events_1000.csv', index=False, encoding='utf-8')
pd.DataFrame(annexes).to_csv('annexes_48.csv', index=False, encoding='utf-8')

# Convert ground truth to DataFrame format
gt_df = []
for user_id, event_ids in ground_truth.items():
    for event_id in event_ids:
        gt_df.append({"user_id": user_id, "event_id": event_id})
pd.DataFrame(gt_df).to_csv('ground_truth_1000.csv', index=False)

print("   âœ“ users_1000.json / csv")
print("   âœ“ events_1000.json / csv")
print("   âœ“ annexes_48.json / csv")
print("   âœ“ ground_truth_1000.json / csv")

# === SAMPLE DATA PREVIEW ===
print("\n" + "="*60)
print("SAMPLE DATA PREVIEW")
print("="*60)

print("\nğŸ‘¤ Sample User:")
sample_user = random.choice(users)
print(f"   ID: {sample_user['id']}")
print(f"   Name: {sample_user['name']}")
print(f"   Wilaya: {sample_user['wilaya']}")
print(f"   Interests: {', '.join(sample_user['interests'])}")

print("\nğŸ‰ Sample Event:")
sample_event = random.choice(events)
print(f"   ID: {sample_event['id']}")
print(f"   Title: {sample_event['title']}")
print(f"   Categories: {', '.join(sample_event['category'])}")
print(f"   Date: {sample_event['date']}")
print(f"   Location: {sample_event['location']}")

print("\nğŸ¢ Sample Annex:")
sample_annex = random.choice(annexes)
print(f"   ID: {sample_annex['annex_id']}")
print(f"   Name: {sample_annex['annex_name']}")
print(f"   Wilaya: {sample_annex['wilaya']}")
print(f"   Activities: {', '.join(sample_annex['activities'])}")

print("\nâœ… Dataset generation complete!")
print("\nğŸ“ Next steps:")
print("   1. Load the data: users_df = pd.read_csv('users_1000.csv')")
print("   2. Train your model with this realistic data")
print("   3. Test recommendation accuracy with ground_truth_1000.json")
```
